Each dataset could have its own specific annotation table. Because of this, I didn't use a for loop; instead, I downloaded each dataset separately

```python
dataset = "GSE69270"
```

```python
gse = GEOparse.get_GEO(geo=dataset, destdir="./data")
```

```python
#download data
expressions = pd.DataFrame()
annotation = pd.DataFrame()
#info_annot =pd.DataFrame(columns=['group','disease state','family','gender','age','tissue', 'GSM'])

print()
print("GSM example:")
for gsm_name, gsm in gse.gsms.items():
    n=0
    info_annot ={}
    print("Name: ", gsm_name)

    df = gse.gsms[gsm_name].table
    df = df.rename(columns={"VALUE": gsm_name})
    df = df.set_index('ID_REF')

    for item in gse.gsms[gsm_name].metadata['characteristics_ch1']:
       key, value = item.split(": ", 1)
       print(value)
       print(key)

       info_annot[key] = value
       info_annot['GSM'] = gsm_name
    # info_annot = info_annot.set_index('GSM')
    # print(info_annot, '\n\n')
    # if expressions.empty:
    if annotation.empty:
       expressions = df
       annotation = pd.DataFrame(info_annot, index=[n])
       n+=1
    else:
       expressions[gsm_name] = df[gsm_name]
       annotation2 = pd.DataFrame(info_annot, index=[n])
       annotation = pd.concat([annotation,annotation2])
       n+=1
```
