# GSE56047

```r
file_path <- "GSE56047/GSE56047-GPL10558_series_matrix.txt"

line_52 <- readLines(file_path, n = 52)[52]
print(line_52)
```

```r
library(stringr)

extract_numbers <- function(string) {
  numbers <- str_extract_all(string, "(?<=racegendersite: )\\b(?:[1-9]|1[0-8])\\b")
  numbers <- unlist(numbers)
  numbers <- as.numeric(numbers)
  return(numbers)
}

extracted_numbers <- extract_numbers(line_52)

print(extracted_numbers)
```

```r
length(extracted_numbers)
```

```r
extracted_numbers[1]
```

```r
gender_map <- setNames(
  c('M', 'M', 'M', 'M', 'M', 'M', 'M', 'M', 'M', 'F', 'F', 'F', 'F', 'F', 'F', 'F', 'F', 'F'),
  c(5, 12, 17, 6, 7, 13, 14, 18, 8, 1, 9, 15, 2, 10, 3, 11, 16, 4)
)

list_of_letters <- sapply(extracted_numbers, function(num) gender_map[[as.character(num)]])

print(list_of_letters)
```

```r
count_GSE56047 <- sum(meta_data$GSE == "GSE56047")
print(count_GSE56047)
```

В таблице meta_data есть столбец GSM и sex, надо взять и добавить к всем GSE56047 M или F согласно списку list_of_letters, то есть в первой строке с GSE56047 добавить первое значение из спискаа  и т.д.  

```r
indices <- which(meta_data$GSE == "GSE56047")

meta_data$sex[indices] <- list_of_letters

meta_data_GSE56047 <- meta_data[indices, ]
print(meta_data_GSE56047)
```
